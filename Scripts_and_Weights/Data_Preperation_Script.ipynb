## Processing Normal Images 

```python
Image_Names = os.listdir(KaggleFilePath)

random.shuffle(Image_Names)
for i in range(196):
    Image = Image_Names[i]
    ImageActualPath = os.path.join(KaggleFilePath,Image)
    ImageTargetPath = os.path.join(TargetPath,Image)
    shutil.copy2(ImageActualPath,ImageTargetPath)
    print("Copying Image : " + str(Image))
```

```python
def TrainTestSplit(ImagesList,Splitsize):
    # split size in decimals 
    NumberofTrainingImages = int(Splitsize*(len(ImagesList)))
    #print(NumberofTrainingImages)
    Train = ImagesList[:NumberofTrainingImages]
    Test = ImagesList[NumberofTrainingImages:]
    return Train,Test
```

```python
#creating Train Test split
CovidImagesPath = r"C:\Users\tejas\Desktop\Machine Learning Codes\Covid_Project\Data\Dataset\Covid"
NormalImagesPath = r"C:\Users\tejas\Desktop\Machine Learning Codes\Covid_Project\Data\Dataset\Normal"

CovidImages = os.listdir(CovidImagesPath)
NormalImages = os.listdir(NormalImagesPath)

print(len(CovidImages),len(NormalImages))

CovidTrain,CovidTest = TrainTestSplit(CovidImages,0.8)
print(len(CovidTrain),len(CovidTest))

NormalTrain,NormalTest = TrainTestSplit(NormalImages,0.8)
print(len(NormalTrain),len(NormalTest))
```

```python
def CreateTrainTestFolders(TrainData,TestData,ActualFilePath):
    Train = "Train"
    Test = "Test"
    TrainPath = os.path.join(ActualFilePath,Train)
    TestPath = os.path.join(ActualFilePath,Test)
    os.mkdir(TrainPath)
    os.mkdir(TestPath)
    print("Processing Train Folder")
    
    for img_names in TrainData:
        ActualPath = os.path.join(ActualFilePath,img_names)
        TargetPath = os.path.join(TrainPath,img_names)
        print("Train Image Processed " + str(img_names) + "...")
        shutil.copy2(ActualPath,TargetPath)
        
    print("Processing Test Folder")
    
    for img_names in TestData:
        ActualPath = os.path.join(ActualFilePath,img_names)
        TargetPath = os.path.join(TestPath,img_names)
        print("Test Image Processed " + str(img_names) + "...")
        shutil.copy2(ActualPath,TargetPath)
    print("Processing Completed ")
```

```python
CreateTrainTestFolders(CovidTrain,CovidTest,CovidImagesPath)
CreateTrainTestFolders(NormalTrain,NormalTest,NormalImagesPath)
```
